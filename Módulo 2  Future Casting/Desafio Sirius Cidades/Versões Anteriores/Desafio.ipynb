```python
import pandas as pd
pd.set_option('display.max_columns', None)
df = pd.read_excel("./Dados/Fonte dos Dados.xlsx")
df.head()
```

```python
df_alunos = pd.read_excel("./Dados/Dados_Projeto_-_Alunos_Sirius.xlsx")
df_alunos.head()
```

```python
df_alunos['Cidade'] = df_alunos['Cidade'].str.lower()
```

```python
df['nomeLocalidade'] = df['nomeLocalidade'].str.lower()
```

```python
# check the values of each row for each column
n = df_alunos.nunique(axis=0)
print("Nº de valores unicos em cada coluna :\n",
      n)
```

```python
# Contar a frequência de cada cidade
cidade_counts = df_alunos['Cidade'].value_counts()

# Criando um DataFrame a partir da série
df_cidade_counts = pd.DataFrame({
    'nomeLocalidade': cidade_counts.index,
    'quantidade de alunos': cidade_counts.values
})
# Adicionando uma coluna com 'Sim' ou 'Não' com base na quantidade de alunos
df_cidade_counts['Cidade tem Aluno'] = ['Sim' if x >= 1 else 'Não' for x in df_cidade_counts['quantidade de alunos']]

df_cidade_counts
```

```python
# Filtrar df com base em df_alunos
df_filtrado = df[df['nomeLocalidade'].isin(df_alunos['Cidade'])]

df_filtrado.head()
```

```python
df_filtrado.info()
```

```python
# Usando a função merge com o índice
df_merged = df_filtrado.merge(df_cidade_counts, on='nomeLocalidade', how='left')
```

```python
df_merged.head()
```

```python
df_merged.corr()["quantidade de alunos"].sort_values()
```

```python
#plotar a distribuição de Produtividade
import seaborn as sns
sns.distplot(df_merged['quantidade de alunos'])
```

```python
df_merged.corr()["quantidade de alunos"].sort_values()
```

```python
correlacao = df_merged.corr() #criando uma matriz de correlação
correlacao_filtrada = correlacao[correlacao["quantidade de alunos"].abs()>0.1]["quantidade de alunos"]
correlacao_filtrada
```

```python
correlacao_filtrada.index
```

```python
df1 = df_merged[correlacao_filtrada.index].copy() #seleção das variáveis de interesse.
```

```python
df1.head()
```

```python
X = df1.drop(["quantidade de alunos"
              ],axis=1)
y = df1["quantidade de alunos"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
```

```python
modelo.fit(X_train, y_train)
```

```python
predicao_modelo = modelo.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(predicao_modelo, y_test)
```

```python
#fazer R2 do modelo
from sklearn.metrics import r2_score
r2_score(y_test, predicao_modelo)
```

```python
from sklearn.feature_selection import RFE
#importar decision tree
from sklearn.tree import DecisionTreeRegressor
#importar knn
from sklearn.neighbors import KNeighborsRegressor

# Create the RFE object and rank each pixel
SVR = DecisionTreeRegressor(random_state=42)
rfe = RFE(estimator=SVR, 
          n_features_to_select=20, 
          step=1)
fit = rfe.fit(X_train, y_train)
```

```python
predicao_rfe = rfe.predict(X_test)
```

```python
mean_absolute_error(predicao_rfe, y_test)
```

```python
r2_score(y_test, predicao_rfe)
```

```python
selected_features = pd.Series(fit.support_, index=X.columns)
print(selected_features[selected_features == True].index)
```
