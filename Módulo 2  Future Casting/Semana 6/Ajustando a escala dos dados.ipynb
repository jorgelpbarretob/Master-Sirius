````python
import pandas as pd
pd.set_option('display.max_columns',None)
df = pd.read_excel("Fonte dos Dados.xlsx")
df.head()
````

````python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
````

````python
df[['Minmax População residente', 'Minmax Densidade demográfica']] =scaler.fit_transform(df[['População residente', 'Densidade demográfica']])
````

````python
 from sklearn.preprocessing import StandardScaler
 scaler = StandardScaler()
 df[['Scaler População residente', 'Scaler Densidade demográfica']] = scaler.fit_transform(df[['População residente', 'Densidade demográfica']])
````
