````python
import pandas as pd
pd.set_option('display.max_columns',None)
df = pd.read_excel("Fonte dos Dados.xlsx")
df.head()
````

````python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
````

````python
df[['Minmax Populao residente', 'Minmax Densidade demogrfica']] =scaler.fit_transform(df[['Populao residente', 'Densidade demogrfica']])
````
